```python
#import sys
#sys.stdin = open('to_replace.txt', 'r')

#i = sys.stdin.read()
f = open('to_replace.txt', 'r', encoding='utf-8')

i = ''.join(line for line in f)

replacements = {
    ' /∈ ' : ' \\notin ',
    '∈' : '\in',
    '×' : '\\times',
    ' 6 = ' : ' \\ne ',
    '⊆' : '\subseteq',
    '{' : '\{ ',
    '}' : ' \}',
    '∅' : '\\varnothing',
    '∩' : '\cap',
    '∪' : '\cup',
    'ˆ' : '\hat ',
    ' 6 % ' : ' \\not\succsim ',
#    ' 6 ' : ' \leqslant ',
    '⊂' : '\subset',
    '%' : '\succsim',
    '' : '\succ',
    'R2\n+' : '\mathbb{R}^2_{+}',
    'RL\n+' : '\mathbb{R}^L_{+}',
    'Rn' : '\mathbb{R}^n',
    '√' : '\sqrt{',
    '−→' : '\\rightarrow',
    '∞' : '\infty',
    'x1' : 'x_1',
    'x2' : 'x_2',
    'xn' : 'x_n',
    '◦' : '\circ',
    'x0' : 'x_0',
    '>̂' : '> \hat',
    '⇐⇒' : '\iff',
    ' e ' : ' \epsilon ',
    'n\inN':'n \in \mathbb{N}',
    'θ' : '\\theta',
    'σ' : '\sigma',
    '→' : ' \\rightarrow ',
    '∑' : '\sum_{',
    '∧' : '\land',
    '=⇒' : '\Rightarrow',
    '∀' : '\\forall',
    '±' : '\pm',
    'ε' : '\epsilon',
    'δ' : '\delta',
}

set_replacements = {
    ' N,' : ' \mathbb{N},',
    ' N.' : ' \mathbb{N}.',
    ' N ' : ' \mathbb{N} ',
    ' Z,' : ' \mathbb{Z},',
    ' Z ' : ' \mathbb{Z} ',
    ' Q,' : ' \mathbb{Q},',
    ' Q ' : ' \mathbb{Q} ',
    ' R,' : ' \mathbb{R},',
    ' R.' : ' \mathbb{R}.',
    ' R ' : ' \mathbb{R} ',
    ' C,' : ' \mathbb{C},',
    ' C ' : ' \mathbb{C} ',
    ' C.' : ' \mathbb{C}.',
}

o = i
for k in replacements.keys():
    o = o.replace(k, replacements[k])

    
sr = True   # Do set notation replacements?
if sr:
    for k in set_replacements.keys():
        o = o.replace(k, set_replacements[k])
    

print('')

print(o)
```
